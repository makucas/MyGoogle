```python
import matplotlib.pyplot as plt
```

```python
with open('executions_time_50.txt', 'r') as arquivo:
    lines = arquivo.readlines()
    values = []
    for line in lines:
        value = float(line.strip())
        values.append(value)
```

```python
plt.plot(values)
plt.title('Tempos de Requisio do Servidor (50rps)')
plt.xlabel('Nmero da Requisio')
plt.ylabel('Tempo (segundos)')
plt.show()

print(f"Nmero total de requisies: {len(values)}")
print(f"Tempo mdio de cada requisio: {sum(values)/len(values):0.2} segundos")
```
