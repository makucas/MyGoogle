```python
import matplotlib.pyplot as plt
```

```python
with open('executions_time_50.txt', 'r') as arquivo:
    lines = arquivo.readlines()
    values = []
    for line in lines:
        value = float(line.strip())
        values.append(value)
```

```python
plt.plot(values)
plt.title('Tempos de Requisição do Servidor (50rps)')
plt.xlabel('Número da Requisição')
plt.ylabel('Tempo (segundos)')
plt.show()

print(f"Número total de requisições: {len(values)}")
print(f"Tempo médio de cada requisição: {sum(values)/len(values):0.2} segundos")
```
